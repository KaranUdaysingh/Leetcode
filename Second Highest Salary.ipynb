```python
def maxProfit(prices):
    """
    :type prices: List[int]
    :rtype: int
    """
    profit=0
                
    if prices==sorted(prices):
        return(prices[-1]-prices[0])
    
    elif prices==sorted(prices,reverse=True):
        return(0)
    
    else:
        temp=prices[0]
        for i in range(1, len(prices)):  
            if temp<prices[i]:
                profit=profit+prices[i]-temp
                temp=prices[i]

            elif temp>prices[i]:
                temp=prices[i]
                continue
    return(profit)
```
